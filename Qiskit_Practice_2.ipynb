```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import array_to_latex
from qiskit_ibm_runtime import Sampler
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import Operator
```

```python
qc = QuantumCircuit(3)
# Apply H-gate to each qubit:
for qubit in range(3):
    qc.h(qubit)
# See the circuit:
qc.draw(output="mpl")
```

```python
final_state = Statevector.from_instruction(qc)

array_to_latex(final_state.data, prefix="\\text{Statevector} = ")
```

```python
qc1 = QuantumCircuit(2)
qc1.h(0)
qc1.x(1)
qc1.draw(output="mpl")
```

```python
unitary = Operator(qc1).data

array_to_latex(unitary, prefix="\\text{Circuit} = ")
```

```python
qc2 = QuantumCircuit(2)
qc2.x(1)
qc2.draw(output="mpl")
```

```python
unitary = Operator(qc2).data

array_to_latex(unitary, prefix="\\text{Circuit} = ")
```

```python
qc5 = QuantumCircuit(2)

qc5.h(0)

qc5.cx(0,1)
qc5.draw(output="mpl")
```

```python
final_state = Statevector.from_instruction(qc5)

array_to_latex(final_state.data, prefix="\\text{Statevector} = ")
```

```python
qc5.measure_all()
```

```python
backend = AerSimulator()

counts = backend.run(qc5, shots=1000).result().get_counts()

print(counts)

plot_histogram(counts)
```
